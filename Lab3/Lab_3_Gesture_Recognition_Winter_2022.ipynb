Defining a few helper functions

```python
def accuracy(output, labels):
    prob_list = F.softmax(output, dim=1)
    index = 0
    accuracy = 0
    for tensor in prob_list:
        temp_list = tensor.tolist()
        # print("---")
        # print(temp_list)
        max_num = 0
        max_index = 0
        for i in range(0, len(temp_list)):
            elem = temp_list[i]
            if elem > max_num:
                max_num = elem
                max_index = i
        if max_index == labels[index]:
            accuracy += 1
        index += 1
        
    return float(accuracy)/index * 100
```

```python
def evaluate(net, loader, criterion):
    """ Evaluate the network on the validation set.

     Args:
         net: PyTorch neural network object
         loader: PyTorch data loader for the validation set
         criterion: The loss function
     Returns:
         err: A scalar for the avg classification error over the validation set
         loss: A scalar for the average loss function over the validation set
     """
    total_loss = 0.0
    total_acc = 0.0
    total_epoch = 0
    for i, data in enumerate(loader, 0):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        total_acc += accuracy(outputs, labels)
        total_loss += loss.item()
        total_epoch += len(labels)
    acc = float(total_acc) / (i + 1)
    loss = float(total_loss) / (i + 1)
    return loss, acc
```

### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
# Defining the CNN architechture
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()

        # 3 input image channel, 6 output channels, 5x5 square convolution
        self.conv1 = nn.Conv2d(3, 32, 5)
        self.conv2 = nn.Conv2d(32, 64, 5)
        self.fc1 = nn.Linear(64 * 53 * 53, 240)  # 53*53 from image dimension
        
        # The convolution decreases dimension by (W - f)/S + 1 and pooling reduces it by a factor of 2
        self.fc2 = nn.Linear(240, 128)

        # The last layer has 9 neurons for each of the 9 categories
        self.fc3 = nn.Linear(128, 9)

    def forward(self, x):
        # Max pooling over a (2, 2) window
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        # If the size is a square, you can specify with a single number
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        x = torch.flatten(x, 1) # flatten all dimensions except the batch dimension
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

net = Net()
net.to(device) # transferring it to the gpu (if available)
print(net)
```

### Part (b) Training Code - 5 pt

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

```python
num_epochs = 20

def train(net, num_epochs, log=True, learning_rate = 0.01, moment = 0.7, loss=False, l2=0):   
    criterion = nn.CrossEntropyLoss()
    if l2:
        optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=moment, weight_decay=l2)
    else:
        optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=moment)
    
    train_loss = np.zeros(num_epochs)
    train_acc = np.zeros(num_epochs)

    val_loss = np.zeros(num_epochs)
    val_acc = np.zeros(num_epochs)

    for epoch in range(num_epochs):  # loop over the dataset multiple times    
        total_train_loss = 0.0
        total_train_err = 0.0
        total_epoch = 0
        total_train_acc = 0.0

        for i, data in enumerate(train_loader, 0):
            # get the inputs; data is a list of [inputs, labels]
            inputs, labels = data
            inputs, labels = inputs.to(device), labels.to(device) # transferring it to the gpu

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward + backward + optimize
            outputs = net(inputs)
            #print(outputs.shape, labels.shape)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            # Calculate the statistics
            total_train_acc += accuracy(outputs, labels)
            total_train_loss += loss.item()
            total_epoch += len(labels)

        # train_err[epoch] = float(total_train_err) / total_epoch
        train_acc[epoch] = float(total_train_acc) / (i+1)
        train_loss[epoch] = float(total_train_loss) / (i+1)
        val_loss[epoch], val_acc[epoch] = evaluate(net, val_loader, criterion)
        
        if log:
            print(("Epoch {:.2f}: Train acc: {:.2f}, Train loss: {:.2f} |"+
                   "Val acc: {:.2f}, Val loss: {:.2f}").format(
                       epoch + 1,
                       train_acc[epoch],
                       train_loss[epoch],
                       val_acc[epoch],
                       val_loss[epoch]))

    print('Finished Training')
    print("Train acc:", train_acc[-1], " - Val acc:", val_acc[-1])
    print("---")
    if not loss:
        return [train_acc, val_acc]
    else:
        return [train_acc, val_acc, train_loss, val_loss]
```

### 4. Transfer Learning [15 pt]
For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal
role in establishing Deep Neural Nets as a go-to tool for image classification problems and we will use an

```python
def accuracy(output, labels):
    prob_list = F.softmax(output, dim=1)
    temp_list = prob_list[0].tolist()

    # print("---")
    # print(temp_list, "LABEL:", labels[0].tolist())
    max_num = 0
    max_index = 0
    for i in range(0, len(temp_list)):
        elem = temp_list[i]
        if elem > max_num:
            max_num = elem
            max_index = i
    if max_index == labels[0].tolist():
        return 1

    #print(temp_list, "<->", max_index, "<->", labels)

    #print(max_index, labels)
    return 0
```

```python
def evaluate(anet, loader, criterion):

    total_loss = 0.0
    total_acc = 0.0
    i = 0
    for elem in loader:
        # print(elem[0].shape, elem[1].shape)
        inputs, labels = elem[0], torch.tensor([elem[1]])
        # inputs, labels = inputs.to(device), labels.to(device)
        
        outputs = anet(inputs)
        loss = criterion(outputs, labels)
        total_acc += accuracy(outputs, labels)
        total_loss += loss.item()
        i += 1

    acc = float(total_acc) / (i+1)
    loss = float(total_loss) / (i+1)
    return loss, acc
```

```python
num_epochs = 100

def train(net, num_epochs, log=True, learning_rate = 0.001, moment = 0.8, loss=False, l2=0):   
    criterion = nn.CrossEntropyLoss()
    if l2:
        optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=moment, weight_decay=l2)
    else:
        optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum = moment)
    
    train_loss = np.zeros(num_epochs)
    train_acc = np.zeros(num_epochs)

    val_loss = np.zeros(num_epochs)
    val_acc = np.zeros(num_epochs)

    for epoch in range(num_epochs):  # loop over the dataset multiple times    
        total_train_loss = 0.0
        total_train_err = 0.0
        total_train_acc = 0.0

        for elem in train_data:
            # get the inputs; data is a list of [inputs, labels]
            inputs, labels = elem[0], torch.tensor([elem[1]])
            # inputs, labels = inputs.to(device), labels.to(device) # transferring it to the gpu

            # forward + backward + optimize
            outputs = net(inputs)

            #print(inputs.shape, outputs.shape, labels.shape)
            
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            # zero the parameter gradients
            optimizer.zero_grad()

        val_loss[epoch], val_acc[epoch] = evaluate(anet, val_data, criterion)
        train_loss[epoch], train_acc[epoch] = evaluate(anet, train_data, criterion)
        
        if log:
            print(("Epoch {:.2f}: Train acc: {:.2f}, Train loss: {:.2f} |"+
                   "Val acc: {:.2f}, Val loss: {:.2f}").format(
                       epoch + 1,
                       train_acc[epoch],
                       train_loss[epoch],
                       val_acc[epoch],
                       val_loss[epoch]))

    print('Finished Training')
    print("Train acc:", train_acc[-1], " - Val acc:", val_acc[-1])
    print("---")
    if not loss:
        return [train_acc, val_acc]
    else:
        return [train_acc, val_acc, train_loss, val_loss]

anet = ANet()
train(anet, num_epochs, log=True, learning_rate = 0.001, moment = 0.8, loss=False, l2=0)
```
